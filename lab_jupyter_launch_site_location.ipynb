```python
# Task 2: Mark the success/failed launches for each site on the map

# Add a marker and circle for each launch site
for index, row in launch_sites_df.iterrows():
    coordinate = [row['Lat'], row['Long']]
    
    # Add a circle at the launch site
    folium.Circle(
        coordinate, 
        radius=1000, 
        color='#000000', 
        fill=True,
        fill_color='#3186cc'
    ).add_child(folium.Popup(row['Launch Site'])).add_to(site_map)
    
    # Add a marker with a label
    folium.map.Marker(
        coordinate,
        icon=DivIcon(
            icon_size=(20,20),
            icon_anchor=(0,0),
            html=f'<div style="font-size: 12px; color:#d35400;"><b>{row["Launch Site"]}</b></div>'
        )
    ).add_to(site_map)

# Define marker cluster
marker_cluster = MarkerCluster().add_to(site_map)

# Iterate through each launch
for index, row in spacex_df.iterrows():
    coordinate = [row['Lat'], row['Long']]
    
    # Define color based on success/failure
    color = 'green' if row['class'] == 1 else 'red'
    
    # Add marker to cluster
    folium.Marker(
        coordinate,
        icon=folium.Icon(color=color, icon="info-sign"),
        popup=f"Launch Site: {row['Launch Site']}<br>Outcome: {'Success' if row['class'] == 1 else 'Failure'}"
    ).add_to(marker_cluster)

site_map
```

```python
# TASK 3: Calculate the distances between a launch site to its proximities


import pandas as pd
import folium
from geopy.distance import geodesic

# Example proximities: coastline, highway, and railway (you can update with real values)
proximities = {
    "Coastline": (28.5634, -80.5670),   # Closest coastline near Cape Canaveral
    "Highway": (28.5721, -80.5853),      # Closest major highway
    "Railway": (28.5729, -80.6545),      # Closest railway
    "Nearest City": (28.3922, -80.6077)  # Example nearest city
}

# Function to calculate distance between two coordinates (lat, long)
def calculate_distance(coord1, coord2):
    return geodesic(coord1, coord2).kilometers

# Create an empty list to store distance results
distance_results = []

# Compute distances from each launch site to proximity points
for index, row in launch_sites_df.iterrows():
    site_name = row['Launch Site']
    site_coord = (row['Lat'], row['Long'])

    for place, place_coord in proximities.items():
        distance = calculate_distance(site_coord, place_coord)
        distance_results.append([site_name, place, distance])

# Convert to DataFrame
distance_df = pd.DataFrame(distance_results, columns=['Launch Site', 'Proximity', 'Distance (km)'])

# Display results



# Initialize the map centered around the first launch site
site_map = folium.Map(location=[launch_sites_df.iloc[0]['Lat'], launch_sites_df.iloc[0]['Long']], zoom_start=10)

# Add launch sites to map
for index, row in launch_sites_df.iterrows():
    folium.Marker(
        location=[row['Lat'], row['Long']],
        icon=folium.Icon(color='blue', icon='rocket'),
        popup=row['Launch Site']
    ).add_to(site_map)

# Add proximity points
for place, place_coord in proximities.items():
    folium.Marker(
        location=place_coord,
        icon=folium.Icon(color='red', icon='info-sign'),
        popup=place
    ).add_to(site_map)

# Display map
site_map

distance_df
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
import folium
from folium.features import DivIcon
from math import sin, cos, sqrt, atan2, radians


# Coordinates of Launch Site (Example: CCAFS SLC-40)
launch_site_lat, launch_site_lon = 28.56342, -80.57674

# Coordinates of Closest Features (Use MousePosition to update these values)
city_lat, city_lon = 28.3922, -80.6077     # Example: Cocoa, Florida
railway_lat, railway_lon = 28.5721, -80.5852   # Closest railway
highway_lat, highway_lon = 28.5639, -80.5707   # Closest highway
coastline_lat, coastline_lon = 28.56367, -80.57163  # Closest coastline

# Calculate Distances
distance_city = calculate_distance(launch_site_lat, launch_site_lon, city_lat, city_lon)
distance_railway = calculate_distance(launch_site_lat, launch_site_lon, railway_lat, railway_lon)
distance_highway = calculate_distance(launch_site_lat, launch_site_lon, highway_lat, highway_lon)
distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)

# Initialize Map
site_map = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=12)

# Function to Add Markers with Distance Labels
def add_marker(lat, lon, label, distance):
    folium.Marker(
        [lat, lon],
        icon=DivIcon(
            icon_size=(20, 20),
            icon_anchor=(0, 0),
            html=f'<div style="font-size: 12px; color:#d35400;"><b>{label}: {distance:.2f} KM</b></div>'
        )
    ).add_to(site_map)

# Add Markers for Closest Features
add_marker(city_lat, city_lon, "City", distance_city)
add_marker(railway_lat, railway_lon, "Railway", distance_railway)
add_marker(highway_lat, highway_lon, "Highway", distance_highway)
add_marker(coastline_lat, coastline_lon, "Coastline", distance_coastline)

# Draw PolyLines from Launch Site to Each Feature
locations = {
    "City": (city_lat, city_lon),
    "Railway": (railway_lat, railway_lon),
    "Highway": (highway_lat, highway_lon),
    "Coastline": (coastline_lat, coastline_lon)
}

for label, (lat, lon) in locations.items():
    folium.PolyLine(
        locations=[[launch_site_lat, launch_site_lon], [lat, lon]],
        color="blue",
        weight=2,
        dash_array="5, 5"
    ).add_to(site_map)

# Display Map
site_map
```
